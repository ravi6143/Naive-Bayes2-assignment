Let's denote the events as follows:

A: Employee uses the health insurance plan.
B: Employee is a smoker.
The probability of an employee using the health insurance plan is denoted as P(A), and the probability of an employee being a smoker is denoted as P(B). The probability that an employee uses the health insurance plan given that they are a smoker is denoted as P(A|B).

From the given information:

P(A)=0.70 (probability of using the health insurance plan),

P(B∣A)=0.40 (probability of being a smoker given using the health insurance plan).

We can use the conditional probability formula to find 

P(B∣A)= P(A).P(A∩B)
​
 

Since P(A∩B) is not directly provided, we can use the fact that P(A∩B)= P(B∣A).P(A):


P(B∣A) = P(B|A).P(A) / P(A)
 

Substitute the values:

P(B∣A)= 0.70 0.40 / 0.70
​
 

Simplify:

P(B∣A)=0.40

Therefore, the probability that an employee is a smoker given that he/she uses the health insurance plan is 
0.40 or 40% .

## Bernoulli Naive Bayes:

* Type of Data: Bernoulli Naive Bayes is suitable for binary or Boolean features, where each feature can take only two possible values (0 or 1).

* Assumption: It assumes that features are binary and follows a Bernoulli distribution, meaning each feature is treated as a binary random variable.


## Multinomial Naive Bayes:

* Type of Data: Multinomial Naive Bayes is designed for discrete data, typically used for text classification where features represent the frequency of terms (word counts) in a document.

* Assumption: It assumes that features follow a multinomial distribution, which is appropriate for count data, such as the number of occurrences of a term in a document.

1. Ignore Missing Values:

In many implementations, missing values are simply ignored during the calculation of probabilities. The presence or absence of a feature (binary) is the focus, and missing values are treated as if the feature is not present.


2. Imputation:

You can choose to impute missing values before applying Bernoulli Naive Bayes. Imputation involves replacing missing values with a specific value, such as the mean, median, or mode of the observed values in that feature.


3. Special Treatment:

In some cases, a special category or value may be assigned to missing values, treating them as a distinct category.

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.33 , random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## Bernoulli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
clf_bernoulli = BernoulliNB()

from sklearn.model_selection import GridSearchCV

parameters = { 'alpha':[1.0], 'force_alpha':[True], 'binarize':[0.0], 'fit_prior':[True], 'class_prior':[None]}

clf_grid_bernoulli = GridSearchCV(BernoulliNB(),param_grid=parameters , cv = 10)

clf_grid_bernoulli.fit(x_train , y_train)
```

```python
y_pred = clf_grid_bernoulli.predict(x_test)
```

```python
from sklearn.metrics import f1_score , accuracy_score , precision_score , recall_score

print(accuracy_score(y_pred , y_test))
print(precision_score(y_pred , y_test))
print(recall_score(y_pred , y_test))
print(f1_score(y_pred , y_test))
```
